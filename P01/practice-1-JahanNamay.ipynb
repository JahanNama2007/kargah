```python
received_msg = "recived_msg.txt"
sent_msg = "sent_msg.txt"
```

```python
with open(received_msg, "r", encoding = "utf-8") as file:
    file_contents = file.read()  #خواندن فایل تکست دریافتی 
```

```python
a=file_contents.splitlines() #تبدیل هر خط فایل دریافتی به یک آرگومان از لیست 
```

```python
# جدا کردن سومین آرگومان لیست که همان اسم شخض است 

NAME = a[2] 
```

```python
with open(sent_msg, "r", encoding = "utf-8") as file:
    file_contents = file.read() #خواندن فایل ارسالی از طرف ما که کارفرما به سایت داده است 
```

```python
sent=file_contents.replace('NAME' , NAME)
# جایگذاری اسم شخص جدا شده  در مرحله  3 به جای 
#"NAME"
# در تکست ارسالی 
```
