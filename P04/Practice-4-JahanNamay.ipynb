```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def pic(image_add):
    img = cv2.imread(image_add)
    rgb_img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)



    


    
    plt.figure(figsize=(12, 10))

    plt.subplot(4,2, 1)
    plt.imshow(rgb_img )
    plt.title('Image')
    plt.axis('off')
    

    plt.subplot(4, 2, 3)
    plt.imshow(gray  , cmap='gray' )
    plt.title('gray Image')
    plt.axis('off')

    
    plt.subplot(4, 2, 4)
    plt.hist(gray.ravel(), bins=256, range=(0, 256))
    plt.title('Histogram of Gray')
    plt.xlim([0, 256])

    
    w2b = gray.copy()
    mask_white=(gray==255)
    mask_black=(gray==0)
    w2b[mask_white]=0
    w2b[mask_black]=255

    
    plt.subplot(4, 2, 5)
    plt.imshow(w2b, cmap='gray')
    plt.title('Black ↔ White')
    plt.axis('off')

    
    plt.subplot(4, 2, 6)
    plt.hist(w2b.ravel(), bins=256, range=(0, 256))
    plt.title('Histogram (Black ↔ White)')
    plt.xlim([0, 256])

    
    inv_img = 255 - gray

    plt.subplot(4, 2,7)
    plt.imshow(inv_img  , cmap='gray')
    plt.title('Inv Image')
    plt.axis('off')

    
    plt.subplot(4, 2, 8)
    plt.hist(inv_img.ravel(), bins=256, range=(0, 256))
    plt.title('Histogram (Inv)')
    plt.xlim([0, 256])

    plt.tight_layout()
    plt.savefig('f.png')
    plt.show()

    
    

pic('file.png')
    
```
